# problem 2


Problem Understanding

Find the sum of all digits in a number.


Logic
* Extract each digit using % 10
* Add it to the sum
* Remove last digit using // 10

```python
""""
Input number
sum = 0

While number > 0
    digit = number % 10
    sum = sum + digit
    number = number // 10
Print sum
"""
```

```python
num = int(input("enter any no"))
sum_digits = 0

while num > 0:
    digit = num % 10
    sum_digits += digit
    num = num // 10

print("Sum of entered nox =", sum_digits)
```

# problem 3


Problem Understanding

Reverse the digits of a number.


Logic
* Take last digit
* Add it to reversed number
* Remove last digit
* Repeat until number becomes 0

```python
"""
Input number
reverse = 0

While number > 0
    digit = number % 10
    reverse = reverse * 10 + digit
    number = number // 10

Print reverse
"""
```

```python
num = int(input("Enter any number: "))
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10
print(reverse)
```
