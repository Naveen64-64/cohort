# PROBLEM 1

PROBLEM UNDERSTANDING 

We need to take the person's age as input and check whether they are eligible to vote.

LOGIC / APPROACH

* Take age from the user.
* If age is 18 or above, the person can vote.
* Otherwise, the person cannot vote.

```python
# Pseudocode

START
INPUT age
IF age >= 18:
    PRINT "Eligible to Vote"
ELSE:
    PRINT "Not Eligible to Vote"
END
```

```python
age = int(input("Enter your age: "))

if age >= 18:
    print("Eligible to Vote")
else:
    print("Not Eligible to Vote")
```

# PROBLEM 2

Problem Understanding

Take student marks and assign grades according to the given criteria.

Logic / Approach

* If marks are 90 or above → Grade A
* If marks are between 75 and 89 → Grade B
* If marks are between 60 and 74 → Grade C
* Otherwise → Fail

```python
# Pseudocode

START
INPUT marks

IF marks >= 90
    PRINT "Grade A"
ELSE IF marks >= 75
    PRINT "Grade B"
ELSE IF marks >= 60
    PRINT "Grade C"
ELSE
    PRINT "Fail"

END
```

```python
marks = int(input("Enter your marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

# PROBLEM 3

Problem Understanding

Calculate the sum of natural numbers from 1 to N using a loop.

Logic / Approach


* Take N as input.
* Start sum with 0.
* Use a loop from 1 to N.
* Add each number to the sum.
* Print the final sum.

```python
# Pseudocode

START
INPUT N
sum = 0

FOR i = 1 TO N
    sum = sum + i

PRINT sum
END
```

```python
n = int(input("Enter a number: "))

total = 0
for i in range(1, n + 1):
    total += i

print("Sum =", total)
```

# problem 4

Problem Understanding

Keep asking the user for a password until the correct password is entered.

Logic / Approach


* Store the correct password.
* Initialize attempts to 0.
* Use a loop.
* Increase attempts for every try.
* Stop when the correct password is entered

```python
# Pseudocode

START

correct_password = "python123"
attempts = 0

WHILE True
    INPUT password
    attempts = attempts + 1

    IF password == correct_password
        PRINT "Access Granted"
        PRINT attempts
        BREAK

END
```

```python
correct_password = "python123"
attempts = 0

while True:
    password = input("Enter Password: ")
    attempts += 1

    if password == correct_password:
        print("Access Granted")
        print("Attempts Taken:", attempts)
        break
    else:
        print("Wrong Password. Try Again!")
```

Problem Understanding

Create an ATM system where users can:

* Check Balance
* Deposit Money
* Withdraw Money
* Exit

* The initial balance is ₹10,000.

Logic / Approach

* Create separate functions:
    * check_balance()
    * deposit()
    * withdraw()
* Use a loop to display the ATM menu repeatedly.
* Update balance after deposits and withdrawals.
* Prevent withdrawing more money than available.

```python
START

balance = 10000

FUNCTION check_balance
    PRINT balance

FUNCTION deposit
    INPUT amount
    balance = balance + amount

FUNCTION withdraw
    INPUT amount

    IF amount <= balance
        balance = balance - amount
    ELSE
        PRINT insufficient balance

WHILE True

    DISPLAY MENU

    INPUT choice

    IF choice == 1
        check_balance()

    ELSE IF choice == 2
        deposit()

    ELSE IF choice == 3
        withdraw()

    ELSE IF choice == 4
        EXIT LOOP

    ELSE
        PRINT invalid choice
END
```

```python
balance = 10000

def check_balance():
    print("Current Balance: ₹", balance)

def deposit():
    global balance
    amount = float(input("Enter deposit amount: ₹"))
    balance += amount
    print("Deposit Successful!")
    print("Updated Balance: ₹", balance)

def withdraw():
    global balance
    amount = float(input("Enter withdrawal amount: ₹"))

    if amount <= balance:
        balance -= amount
        print("Withdrawal Successful!")
        print("Remaining Balance: ₹", balance)
    else:
        print("Transaction Failed!")
        print("Insufficient Balance!")

while True:
    print("\n===== ATM MENU =====")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        check_balance()

    elif choice == 2:
        deposit()

    elif choice == 3:
        withdraw()

    elif choice == 4:
        print("Thank you for using the ATM!")
        break

    else:
        print("Invalid Choice! Please Try Again.")
```
